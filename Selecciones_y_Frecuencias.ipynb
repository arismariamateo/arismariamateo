##Selecciones y Frecuencia

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
# Filtrar: Seleccione solamente los inmuebles clasificados con tipo "Departamento" 
```

```python
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion]
n1
#la frecuencia es 19532 las vece que se repite las rows
```

```python
#la frecuencia es 19532 las veces que se repite las rows
n1 = datos[seleccion].shape[0]
n1
```

```python
#Seleccione los inmuebles clasificados con tipo de "Casa", "Casa en condominio" y "Casa en Villa"
```

```python
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa en villa')
n2 = datos[seleccion].shape[0]
n2
```

```python
#Selecciones los inmuebles con áreas entre 60 y 100 metro cuadrados, incuyendo lo límites
```

```python
seleccion = (datos['Area'] >= 60) & (datos['Area'] <=100)
n3 = datos[seleccion].shape[0]
n3
```

```python
#Selecciones los inmuebles que tengan por lo menos cuatro cuartos y alquiler menor de $2.000,00
```

```python
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] <2000)
n4 = datos[seleccion].shape[0]
n4
```

```python
print('No. de inmuebles clasificados con tipo "Departamento" '.format(n1))
print('No. de inmuebles clasificados con tipo de "Casa", "Casa en condominio" y "Casa en Villa" '.format(n2))
print('No. de inmuebles con áreas entre 60 y 100 metro cuadrados, incuyendo lo límites '.format(n3))
print('No. de inmuebles que tengan por lo menos cuatro cuartos y alquiler menor de $2.000,00 '.format(n4))
```
